Ce programme simule le processus de Cannings, 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import collections as mc
```

```python
# parameters
N = 3
```

```python
M = 2 * N
ancestor = np.arange(M)
all_parents = []

while not all(ancestor == ancestor[0]):
    parents = np.random.choice(M, M)
    parents.sort()
    ancestor = ancestor[parents]
    all_parents.append(parents)
all_parents = np.array(all_parents)
```

```python
n_generations = all_parents.shape[0]

points = [(x,y+1) for x in range(n_generations+1) for y in range(M)]

lines = []
for gen in range(n_generations):
    for i in range(M):
        lines.append([points[(gen+1)*M+i], (gen,all_parents[gen,i]+1)])
```

```python
lc = mc.LineCollection(lines, colors='black', linewidths=2)
fig, ax = plt.subplots()
ax.add_collection(lc)   
ax.autoscale()      
ax.margins(0.1)
plt.scatter(*zip(*points), c='black')
```
